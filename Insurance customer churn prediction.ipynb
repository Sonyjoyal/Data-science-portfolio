```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

```

```python
df=pd.read_csv('insurance_data.csv')  #classifcation :-yes or no type
df.head()
```

```python
xpoints=df['age']
ypoints=df['bought_insurance']
```

```python
plt.scatter(xpoints,ypoints,marker='*',color='red')
plt.xlabel('age')
plt.ylabel('bought insurance')
plt.show()
```

```python
x=df[['age']]
y=df['bought_insurance']
```

```python
model=LogisticRegression()
model.fit(x,y)
```
